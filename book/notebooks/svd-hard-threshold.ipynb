```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.diag(
    np.concatenate((
        [1.7, 2.5],
        np.zeros(98)
    ))
)

Y = X + np.random.randn(100, 100) / np.sqrt(100)

plt.imshow(Y)
plt.show()
```

```python
U, S, Vh = np.linalg.svd(Y)

plt.scatter(np.arange(len(S)), S, s = 5)
plt.grid()
plt.show()
```

```python
plt.hist(S, bins = 15, edgecolor = 'black')
plt.show()
```
