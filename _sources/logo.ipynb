```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x : 1 / (1 + np.exp(-x))
df = lambda x : f(x) * (1 - f(x))

x = np.linspace(-5, 5, 100)
plt.figure(facecolor = "white")
plt.plot(x, f(x), linewidth = 5)
plt.plot(x, df(x), linewidth = 3, zorder = 0)
plt.text(-3.5, 0.3, "$\sigma \circ \Sigma$", fontsize = 90, color = "black")
plt.axis(False)
plt.show()
```
